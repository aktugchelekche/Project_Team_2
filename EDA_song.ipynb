```python
# Import dependencies
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import psycopg2

from sqlalchemy import create_engine

from config import db_password
```

```python
# Read the songs_normalize.csv

songs1_df = pd.read_csv("Resources/songs_normalize_1.csv")
songs2_df = pd.read_csv("Resources/songs_normalize_2.csv")

display(songs1_df)
display(songs2_df)
```

```python
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/songs_data"
```

```python
#Create the database engine with the following:
engine = create_engine(db_string)
```

```python
# Find null values
for column in songs1_df.columns:
    print(f"Column {column} has {songs1_df[column].isnull().sum()} null values")
```

```python
songs1_df.to_sql(name='songs1', con=engine, if_exists='replace')
songs2_df.to_sql(name='songs2', con=engine, if_exists='replace')
```

```python
# Joining songs1 and songs2 tables
SELECT songs1_df, songs2_df
FROM songs1
FULL OUTER JOIN songs
ON song = song;
```
